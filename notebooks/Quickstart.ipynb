## 1) Install

```python
import sys

# Install uv into this kernel's environment first: a bare `!uv` needs uv on
# PATH, which a reader running this notebook from a plain Python will not have.
!{sys.executable} -m pip install --quiet --upgrade uv

# install igittigitt from GitHub (or use: igittigitt from PyPI)
!{sys.executable} -m uv pip install --python {sys.executable} --upgrade git+https://github.com/bitranox/igittigitt.git
```

## 2) Library: ignore mode with a negation

```python
import igittigitt

parser = igittigitt.IgnoreParser()
parser.add_rule("*.py[cod]", base_path="/home/user/project")
parser.add_rule("!keep.pyc", base_path="/home/user/project")

print("main.pyc ignored:", parser.match("/home/user/project/main.pyc"))
print("keep.pyc ignored:", parser.match("/home/user/project/keep.pyc"))
```

## 3) Library: include / whitelist mode

```python
inc = igittigitt.IncludeParser()
inc.add_rule("*.py", base_path="/home/user/project")
print("deep .py kept:", inc.match("/home/user/project/pkg/deep.py"))
```
